```python
import numpy as np
import pandas as pd
import random
import os
from py7zr import SevenZipFile
```

```python
your_name = 'malak'
your_folder = 'ImageAndText_dir'
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
csv_path = f'/content/gdrive/MyDrive/{your_folder}/OCR Project/Datasets/Batch to Generate/Generation_batches/no_tashkeel_{your_name}_batch.csv'
```

```python
mybatch = pd.read_csv(f'{csv_path}')
mybatch
```

```python
generation_list = mybatch['text'].tolist()
```

```python
len(generation_list)
```

```python
baghdad_list = generation_list[0:52500]
rakkas_list = generation_list[52500:105000]
bahij_list = generation_list[105000:]
```

```python
baghdad_generator = GeneratorFromStrings(baghdad_list,
                              count=len(baghdad_list),
                              size=75,
                              background_type=1,
                              fit=True,
                              fonts=['baghdad.ttf'],
                              rtl = True)

os.mkdir(f'{your_name}_generated_baghdad')
path = f"./{your_name}_generated_baghdad/{your_name}_no_tashkeel_baghdad"

for i, (img, lbl) in enumerate(baghdad_generator):
    image_info = img.info.copy()
    image_info['dpi'] = (300, 300)
    img_path = f"{path}_{i+1}.tif"
    txt_path = f"{path}_{i+1}.gt.txt"

    img.save(img_path, **image_info)
    with open(txt_path, 'w') as f:
        f.write(lbl)

    # Print statement to indicate saved files
    print(f"Saved image: {img_path}")
    print(f"Saved label: {txt_path}")
```
