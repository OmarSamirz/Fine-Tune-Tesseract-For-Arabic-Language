```python
from trdg.generators import GeneratorFromStrings

# Ensuring all text entries are strings
text = ['       ', ', , , , , , , , , ',
        '.!:<>()@', ' ', '12323434534565679790',
      '     ','        ']

print(type(text))

# Convert all items to strings explicitly
text = [str(t) for t in text]

# Print each item to check
#for item in text:
    #print(type(item), item)

# Generator setup
generator = GeneratorFromStrings(
    text,
    count=len(text),
    background_type=1,
    fonts=['baghdad.ttf'],
    size=100,
    rtl=True,
)

# Generating and displaying images
for idx, (img, lbl) in enumerate(generator):
    if idx == len(text):
        break
    display(img)
```
